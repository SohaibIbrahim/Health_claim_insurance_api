# Health Insurance Claim Api

##### This is a Health insurance claim api built using Fast api,which can fetch,store,update and delete insurance claims and also implements user authentication

##### File structure of this api:

* models : Defines the database aspects of the api
* routers : Creates and manages api endpoints(get,post,delete....)
* schemas : Used for data validation(Whether the data given by the user is acceptable or not)
* services : Overall logic of the api
* core : This is used for configuration and security implementations in this api

# Working of api

- the api is started with uvicorn
- the user is required to signup(account credentials are hashed and stored)
- the user then needs to sign in with the creddentials
- credentials are in form of tokens generated by python-jose 
- jose uses jwt tokens for creating and verifying the tokens
- it then verifies and returns info or returns 401 unauthorized error
- User then have to create a claim using post
- then they have to get claim with GET
- claim is then referred with a claim_if
- Then the user can get status **APPROVER**
                               **DENIED**
                               **PENDING**
- After this the claim can be deleted or stored in Sqlite claims.db file in root folder
